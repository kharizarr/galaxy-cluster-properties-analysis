```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as pl
```

```python
all = pd.read_csv('abell2124_tess.csv')
sample = pd.read_excel('abell2124_dr18 olahan.xlsx', sheet_name='data bersih')
sample_red = pd.read_excel('abell2124_tess.xlsx', sheet_name='abell2124_tess')
bukhari = pd.read_excel('anggota 2 metode.xlsx', sheet_name='Bukhari')
zoom = sample[(sample['z'] < 0.2)]
cc = pd.read_excel('Color cut_dr18.xlsx', sheet_name='anggota hasil irisan')
cc_red = pd.read_excel('Color cut_dr18.xlsx', sheet_name='anggota hasil irisan (2)')
irisan = pd.read_excel('Hasil irisan dan massa.xlsx', sheet_name='Sheet1')
gab = pd.read_excel('Hasil gabungan dan massa.xlsx', sheet_name='Sheet1')
```

```python
center = bukhari[(bukhari['ra'] > 236.24) & (bukhari['ra'] < 236.25) & (bukhari['dec'] > 36.10) & (bukhari['dec'] < 36.11)]
```

```python
plt.subplots(figsize=(6, 5))
plt.scatter(all['ra'],all['dec'], color='grey', marker='.', alpha=0.3, label='Non Anggota')
plt.scatter(cc['ra'],cc['dec'], color='green', marker='.', alpha=1,label='Color Cut')
plt.scatter(bukhari['ra'],bukhari['dec'], color='orange', marker='.', alpha=1, label='Bukhari & Cram')
plt.scatter(irisan['ra'],irisan['dec'], color='red', marker='.', alpha=1, label='Irisan')
#plt.scatter(center['ra'],center['dec'], marker='x', color='red')
plt.xlabel("RA ($^o$)")
plt.ylabel("Dec ($^o$)")
circle = plt.Circle((236.24598, 36.10946), 0.63, fill=False, color='red')
plt.gca().add_patch(circle)
plt.gca().set_xlim((235.45, 237))
plt.gca().set_ylim((35.45, 36.77))
plt.legend()
plt.show()
```
