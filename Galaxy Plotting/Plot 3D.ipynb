```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as pl
import plotly.graph_objs as go
```

## Plot 3D ##

```python
sample = pd.read_excel('abell2124_dr18 olahan.xlsx', sheet_name='data bersih')
bukhari = pd.read_excel('anggota 2 metode ver 2.xlsx', sheet_name='Bukhari')
zoom = sample[(sample['z'] < 0.2)]
cc = pd.read_excel('Color cut_dr18.xlsx', sheet_name='anggota hasil irisan (2)')
cc_zoom = cc[(cc['z'] < 0.1)]
irisan = pd.read_excel('Hasil irisan dan massa.xlsx', sheet_name='Sheet1')
photoz = pd.read_excel('abell2124_dr18 olahan.xlsx', sheet_name='Hasil')
pca = pd.read_excel('sampel pca.xlsx', sheet_name='Sheet7')
gab = pd.read_excel('Hasil gabungan dan massa.xlsx', sheet_name='Sheet1')
gab_zoom = gab[(gab['z'] < 0.2)]
```

```python
fig = go.Figure(data=go.Scatter3d(
    x=zoom['ra'],
    y=zoom['dec'],
    z=zoom['z'],
    mode='markers',
    marker=dict(
        size=4,
    )
    ,
    name='Lingkungan A2124'
))

fig.add_trace(go.Scatter3d(
    x=bukhari['ra'],
    y=bukhari['dec'],
    z=bukhari['z'],
    mode='markers',
    marker=dict(
        size=4,
        color='orange',              
    ),
    name='Anggota Bukhari & Cram'
))
fig.show()
```
